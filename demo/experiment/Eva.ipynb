```python
f=open('shelloutput_V1')
content=f.readlines()
```

```python
trainLoss112=[]
trainLoss224=[]
trainLoss560=[]
trainMAP112=[]
trainMAP224=[]
trainMAP560=[]
testLoss=[]
testMAP=[]
best=[]
for l in content:
    if 'Scale: 112' in l:
        pos=l.index('trainLoss')
        if pos==-1:
            print('Wrong1')
            break;
        s=l[pos+10:pos+16] # right boundary is not included
        trainLoss112.append(float(s))
        
        pos=l.index('trainMAP')
        if pos==-1:
            print('Wrong2')
            break;
        s=l[pos+9:]
        trainMAP112.append(float(s))
        
    if 'Scale: 224' in l:
        pos=l.index('trainLoss')
        if pos==-1:
            print('Wrong1')
            break;
        s=l[pos+10:pos+16] # right boundary is not included
        trainLoss224.append(float(s))
        
        pos=l.index('trainMAP')
        if pos==-1:
            print('Wrong2')
            break;
        s=l[pos+9:]
        trainMAP224.append(float(s))
        
    if 'Scale: 560' in l:
        pos=l.index('trainLoss')
        if pos==-1:
            print('Wrong1')
            break;
        s=l[pos+10:pos+16] # right boundary is not included
        trainLoss560.append(float(s))
        
        pos=l.index('trainMAP')
        if pos==-1:
            print('Wrong2')
            break;
        s=l[pos+9:]
        trainMAP560.append(float(s))
        
    if 'testLoss ' in l:
        s=l[9:15]
        testLoss.append(float(s))
        
        pos=l.index('testMAP')
        s=l[pos+8:pos+14]
        testMAP.append(float(s))
        
    if 'best' in l:
        s=l[10:]
        best.append(float(s))
        
testLoss112=testLoss[0:20]
testLoss224=testLoss[20:40]
testLoss560=testLoss[40:60]
testMAP112=testMAP[0:20]
testMAP224=testMAP[20:40]
testMAP560=testMAP[40:60]
best112=best[0:20]
best224=best[20:40]
best560=best[40:60]
```
